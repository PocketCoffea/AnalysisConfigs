```python
jet_pt = ak.flatten(events.Jet.pt)
jet_eta = ak.flatten(events.Jet.eta)
jet_phi = ak.flatten(events.Jet.phi)
jet_partonFlavour = ak.flatten(abs(events.Jet.partonFlavour))

print("Jet pt:", jet_pt, len(jet_pt))
jet_kin = {
    r"$p_{T}^{jet}$ [GeV]": (jet_pt,  "log", 50, (0, 3000)),
    r"$\eta^{jet}$": (jet_eta, "linear", 50, (-5, 5)),
    r"$\phi^{jet}$": (jet_phi, "linear", 50, (-3.5, 3.5)),
    r"$partonFlavour^{jet}$": (jet_partonFlavour,  "log", 21, (1, 22))
}
```

```python
for var, data in jet_kin.items():
    fig, ax = plt.subplots()
    plt.hist(data[0],  histtype="step", linewidth=2, density=True, bins=data[2], range=data[3])
    plt.xlabel(var)
    plt.ylabel("Density")
    plt.yscale(data[1])
    hep.cms.label(
            year="2022",
            com="13.6",
            # label=f"Private Work ({correct_eta_bins[i]} <"
            # + r"$\eta^{Gen}$"
            # + f"< {correct_eta_bins[i+1]})",
            label=f"Private Work",
            ax=ax,
        )
    plt.show()
    # plt.savefig(f"plots/qcd_{var}.png")
    # plt.close()
    # print(f"Saved plots/qcd_{var}.png")
```
