```python
# plot the JetGood_pt
plt.figure()
plt.hist(jet_pt, bins=500, range=(0, 500), histtype="step", color="blue")
plt.hist(jet_pt_wide, bins=500, range=(0, 500), histtype="step", color="red")

plt.xlabel("JetGood_pt")
plt.ylabel("Events")
# plt.yscale("log")
plt.show()
```

```python
def normalize(array_input):
    mean = np.mean(array_input)
    return (array_input - mean) / np.std(array_input)
```

```python
jet_pt_norm = normalize(jet_pt)
jet_pt_wide_norm = normalize(jet_pt_wide)
plt.figure()
plt.hist(jet_pt_norm, bins=500, range=(-3, 3), histtype="step", color="blue")
plt.hist(jet_pt_wide_norm, bins=500, range=(-3, 3), histtype="step", color="red")
plt.yscale("log")

plt.show()

print(np.mean(jet_pt_norm), np.std(jet_pt_norm))
print(np.mean(jet_pt_wide_norm), np.std(jet_pt_wide_norm))
```

```python
jet_pt_wide_log=np.log(jet_pt_wide+1)
jet_pt_log=np.log(jet_pt+1)
```
